```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd
books = pd.read_csv('books_cleaned.csv')
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_descriptions.txt", sep = "\n" ,index=False, header=False)
```

```python
raw_documents = TextLoader("tagged_descriptions.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size = 0, chunk_overlap = 0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
db_books = Chroma.from_documents(documents, HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2"))
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book related to business ")
```
